# Matplotlib 实践

## 1. matplotlib图表的组成元素

### 1.1 主要元素

#### Axis实例

Axis实例包含：
1. 坐标轴: xaxis, yaxis
2. 刻度: xlim, ylim
3. 标签: xlabel, ylabel
4. 线: grid
5. 标记: annotate

### 1.3 函数用法

#### 1.3.1 函数plot() -- 趋势图：展现变量的趋势变化


```python
x = np.linspace(0.05, 10, 1000)
y = np.cos(x)

plt.plot(x, y, ls='-', lw=2, label='plot figure')
plt.legend()
plt.show()
```

#### 1.3.2 函数scatter() -- 散点图：展现变量之间的关系

```python
x = np.linspace(0.05, 10, 1000)
y = np.random.rand(1000)
plt.scatter(x, y, label='scatter figure')

plt.legend()
plt.show()
```

#### 1.3.3 函数xlim() -- X范围：设置X轴的数值显示范围

```python
plt.scatter(x, y, label = 'scatter figure')
plt.xlim(0.05, 10)
plt.ylim(0,1)

plt.show()
```

#### 1.3.4 函数xlabel -- X标签：设置x轴的标签文本

```python
x = np.linspace(0.05, 10, 1000)
y = np.sin(x)

plt.plot(x, y, ls='-.', lw=1, c='r', label="plot figure")
plt.legend()

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

#### 1.3.5 函数grid()--网格：绘制刻度线的网格线

```python
plt.plot(x, y, ls='-.', lw=2, c='c', label="plot figure")
plt.legend()
plt.grid(linestyle=":", color='r')
plt.show()
```

#### 1.3.6 函数axhline(),axvline()--参考线: 绘制平行于坐标轴的参考线

```python
plt.plot(x, y, ls="-.", lw=2, c='c', label='plot figure')
plt.legend()
plt.axhline(y=0.0, c='r', ls='--', lw=2)
plt.axvline(x=4.0, c='r', ls='--', lw=2)

plt.show()
```

#### 1.3.7 函数axhspan(),axvspan()--绘制垂直于坐标轴的参考区域

```python
plt.plot(x, y, ls='-.', lw=2, c='c', label='plot figure')
plt.axvspan(xmin=4.0, xmax=6.0, facecolor='y', alpha=0.3)
plt.axhspan(ymin=0.0, ymax=0.5, facecolor='y', alpha=0.3)

plt.legend()
plt.show()
```

#### 1.3.8 函数annotate()--注解：添加图形内容细节的指向型注释文本

```python
plt.plot(x,y, ls='-.', lw=2, c='c', label='plot figure')

plt.legend()
plt.annotate('maximum', 
             xy = (np.pi/2, 1.0),
             xytext=((np.pi/2)+1.0, 0.8), 
             weight='bold', color='b',
             arrowprops=dict(arrowstyle='-|>', 
                             connectionstyle='arc3',
                             color='b'))
plt.show()
```

#### 1.3.9 函数text()--文本：添加图形内容细节的无指向型注释文本

```python
plt.plot(x,y, ls='-.', lw=2, c='c', label='plot figure')

plt.legend()
plt.text(3.10, 0.09, 'y=sin(x)', weight='bold', color='r')
plt.show()
```

#### 1.3.10 函数text()--标题：添加图形内容的标题

```python
plt.plot(x,y, ls='-.', lw=2, c='c', label='plot figure')

plt.legend()
plt.title('y=sin(x)')
plt.show()
```

#### 1.3.11 函数legend()--图例：标示不同图形的文本标签图例

```python
plt.plot(x,y, ls='-.', lw=2, c='c', label='plot figure')

plt.legend(loc='upper right')

plt.show()
```

### 1.4 组合应用

下列代码绘制展示了图表组成元素的图表

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm

plt.figure(figsize=(12, 8))

# 定义数据
x = np.linspace(0.5, 3.5, 100)
y = np.sin(x)

y1 = np.random.randn(100)

# 散点图
plt.scatter(x, y1, c='0.25', label='scatter figure')

# 趋势图
plt.plot(x, y, ls='--', lw=2, label='plot figure')


# 设置坐标轴范围
plt.xlim(0.0, 4.0)
plt.ylim(-3.0, 3.0)

plt.ylabel('y axis')
plt.xlabel('x axis')

# 设置坐标网格
plt.grid(True, ls=':', color='r')

# 添加水平线
plt.axhline(y=0.0, c='r', ls='--', lw=2)

# 添加水平区域
plt.axvspan(xmin=1.0, xmax=2.0, facecolor='y', alpha = .3)

# 设置注解
plt.annotate('maximum', xy=(np.pi/2, 1.0),
            xytext=((np.pi/2)+0.15, 1.5), weight='bold', color='r',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='r'))

plt.annotate('spines', xy=(0.75, -3),
            xytext=(0.35, -2.25), weight='bold', color='b',
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='b'))

plt.annotate('', xy=(0,-2.78), xytext=(0.4, -2.32),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='b'))

plt.annotate('', xy=(3.5, -2.98), xytext=(3.6, -2.70),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='b'))

# 设置文本信息
plt.text(3.6, -2.70, '"|" is tickline', weight='bold', color='b')
plt.text(3.6, -2.95, '3.5 is ticklabel', weight='bold', color='b')

plt.title('Structure of matplotlib figure')

plt.legend()

plt.show()
```

## 2. 统计函数简答绘图

### 2.1 函数bar()--柱状图

```python

import matplotlib.pyplot as plt
# 配置参数
# plt.rcParams['font.family'] = ['sans-serif']
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 简单数据
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 1, 4, 5, 8, 9, 7, 2]

plt.bar(x, y, align='center', color='c', tick_label=['q','a','c','e','r','j','b','p'], hatch='/')

plt.xlabel("箱子编号")
plt.ylabel('箱子重量（kg）')

plt.show()
```

### 2.2函数barh()--条形图 

```python
plt.barh(x, y, align='center', color='c', tick_label=['q','a','c','e','r','j','b','p'], hatch='/')

plt.xlabel("box number")
plt.ylabel('weight:kg')

plt.show()
```

### 2.3函数hist()--直方图

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(0, 10, 100)
y = range(0, 11, 1)

plt.hist(x,y, color='g', histtype='bar', rwidth=2, alpha=0.6)
plt.xlabel('box weight(kg)')
plt.ylabel('quantity')

plt.show()
```

### 2.4 函数pie()--饼图

```python
kinds = 'Simple', 'vacuum', 'baggage', 'seal'
colors = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3']
percentages = [3, 4, 5, 6]

plt.pie(percentages, labels=kinds, autopct='%3.1f%%', startangle=60, colors=colors)
plt.show()
```

### 2.5 函数polar()--极线图

```python
theta = np.linspace(0.0, 2 * np.pi, 12, endpoint=False)
r = 30 * np.random.rand(12)

plt.polar(theta, r, color='y', linewidth=1, marker='*', mfc='r', ms=12)
plt.show()
```

### 2.6 函数scatter()--气泡图

```python
import matplotlib as mpl

x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x, y, s=np.power(10 * x + 20 * y, 2), 
            c= np.random.rand(100),
            cmap=mpl.cm.RdYlBu,
            marker='o')
plt.show()
```

### 2.7 函数stem()--棉棒图

```python
x = np.linspace(0.5, 2 * np.pi, 20)
y = np.random.randn(20)

plt.stem(x, y, linefmt='-.', markerfmt='o', basefmt='-', use_line_collection=True)
plt.show()
```

### 2.8 函数boxplot()--箱线图

```python
x = np.random.randn(1000)

plt.boxplot(x)

plt.xticks([1], ['random generator'])
plt.ylabel('random values')

plt.title('stablity of random generator')

plt.grid(axis='y', ls=':', lw=1, color='gray', alpha=0.4)
plt.show()
```

### 2.9 函数errorbar()--误差棒图

```python
x = np.linspace(0.1, 0.6, 6)
y = np.exp(x)

plt.errorbar(x, y, fmt='bo:', yerr=0.2, xerr=0.02)
plt.xlim(0, 0.7)

plt.show()
```

## 3. 绘制统计图形

### 3.1 柱状图

```python
import numpy as np
import matplotlib.pyplot as plt

x = range(1, 6)
y = np.random.randint(1, 10, 5)

plt.bar(x, y, align='center', color='b', tick_label=[chr(i) for i in range(ord('A'),ord('F'))], alpha=0.6)
plt.xlabel('difficulty')
plt.ylabel('copies')

plt.grid(True, axis='y', ls=':', color='r', alpha=0.3)

plt.show()
```

```python
# 绘图案例 an example of matplotlib
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn
from IPython.display import display, clear_output
import time
x = np.linspace(0,5)
f, ax = plt.subplots()
ax.set_title("Bessel functions")

for n in range(1,10):
    time.sleep(1)
    ax.plot(x, jn(x,n))
    clear_output(wait=True)
    display(f)

# close the figure at the end, so we don't get a duplicate
# of the last plot
plt.close()
```

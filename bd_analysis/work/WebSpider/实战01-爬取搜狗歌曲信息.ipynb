# 01 实战-爬取搜狗歌曲信息

## 代码实现

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.kugou.com/yy/rank/home/1-8888.html"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh;\
Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) \
Chrome/75.0.3770.100 Safari/537.36'}
```

```python
r = requests.get(url, headers = headers)
```

```python
r.encoding = 'utf-8'
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
music_div = soup.find('div', class_='pc_temp_songlist')
music_list = music_div.find_all('li')
musics = []
for music_li in music_list:
    rank = music_li.find('span', class_='pc_temp_num')
    rank = rank.text.strip()
    title = music_li.get('title')
    artist = title.split('-')[0].strip()
    title = title.split('-')[1].strip()
    time = music_li.find('span', class_='pc_temp_time')
    time = time.text.strip()
    music_info = {'rank': rank, 
                  'artist': artist, 
                  'title': title, 
                  'time': time
                 }
    musics.append(music_info)
```

```python
import csv
filename = 'sougou.csv'
data = musics
with open(filename, 'w', encoding='utf-8') as f:
    fieldnames = data[0]
    dict_writer = csv.DictWriter(f, fieldnames)
    dict_writer.writeheader()
    dict_writer.writerows(data)
```

#### 2. POST请求

```python
login_form_data = {'username': 'admin', 'password': '123456'}
r = requests.post('https://httpbin.org/post', data = login_form_data)
print(r.text)

# 当响应码是4XX/5XX时，主动抛出异常
r.raise_for_status()
```

#### 3. Cookie处理

```python
user_agent = 'Mozilla/4.0 (compatible; MSIE 5.5; Windows NT)'
headers = {'User-Agent': user_agent}

r = requests.get('http://www.baidu.com', headers=headers)
for cookie in r.cookies.keys():
    print(cookie + ':' + r.cookies.get(cookie))
```

```python
r.headers
```

```python
cookies = dict(name='admin', session_id='100010001')
r = requests.get('http://httpbin.org/cookies', headers=headers, cookies=cookies)
print(r.text)
```
